The graph above illustrates the Bias–Variance Tradeoff. As model complexity increases:

High Bias (Underfitting):
The model is too simple.
It cannot capture important patterns in the data.
Both training and testing errors are high.
High Variance (Overfitting):
The model is too complex.
It memorizes the training data, including noise.
Training error is very low, but testing error becomes high.
Optimal Model:
Lies between underfitting and overfitting.
Provides the lowest testing (generalization) error by balancing bias and variance.
Suppose you are building a machine learning model that predicts Zomato restaurant ratings based on features such as restaurant type, cuisine, location, average cost, online delivery, table booking, and number of reviews.

If the model has high bias, it will be too simple and fail to learn the relationship between these features and restaurant ratings. As a result, it may predict similar ratings for many restaurants, leading to underfitting and poor accuracy.
If the model has high variance, it will fit the training data too closely, even learning random fluctuations or unusual reviews. It may predict the training data very accurately but perform poorly on new restaurants, resulting in overfitting.
The best predictor maintains a balance between bias and variance. Such a model captures meaningful patterns from the training data while still making accurate predictions for new restaurants that it has never seen before. This balance leads to better recommendation quality and more reliable rating predictions.

```python
# session 3 question 5
from sklearn.preprocessing import LabelEncoder
import numpy as np

roles = ['batsman', 'bowler', 'allrounder', 'wicketkeeper']

df['player_role'] = np.random.choice(roles, size=len(df))


le = LabelEncoder()

df['player_role_encoded'] = le.fit_transform(df['player_role'])
print(df[['player_role', 'player_role_encoded']].head())
print("mapping:")
for role, code in zip(le.classes_, range(len(le.classes_))):
    print(f"{role}{code}")
```

```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import RFE

df = pd.read_excel("/content/zomato_small.xlsx")
print(df.head())

le = LabelEncoder()
df['online_order'] = le.fit_transform(df['online_order'])
df['book_table'] = le.fit_transform(df['book_table'])

x = df[['votes', 'online_order', 'book_table', 'rating']]
y = df['average_cost_for_two']

model = DecisionTreeRegressor(random_state=42)
selector = RFE(estimator=model, n_features_to_select=3)

selector.fit(x, y)

print(x.columns[selector.support_])

ranking = pd.DataFrame({
    'Feature': x.columns,
    'Ranking': selector.ranking_
})

print(ranking)
```
